```python
import ast
import nltk 
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
question=[]
answer=[]
```

```python
with open('qa_Electronics.json','r') as f:
    for line in f:
        data=ast.literal_eval(line)
        question.append(data['question'].lower())
        answer.append(data['answer'].lower())
```

```python
question=pd.Series(question)
question
```

```python
answer
```

```python
vectorizer=TfidfVectorizer(norm='l1')
tf_idf_matrix=vectorizer.fit_transform(question)
```

```python
def conversation(im):
    Y_vec=vectorizer.transform(im)
    angle=max(cosine_similarity(Y_vec,tf_idf_matrix)[0])
                     
    if angle<0.6:
        return "Sorry, I did not quite understand that"
    else:
        return answer[np.argmax(cosine_similarity(tf_idf_matrix,Y_vec))]
```

```python
def main():
    user = input("Enter your Username: ") 
    print(f"Welcome to tech support {user}")
    while True:
        data=input(f"\n{user}:")
        if data=='bye':
            break
        else:
            g=conversation([data])
            print(f"\nQ&A Support:{g}")
main()
```
